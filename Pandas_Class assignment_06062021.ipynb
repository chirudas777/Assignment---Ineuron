# Some analysis to answer some questions

```python
#Which year had the highest total crime numbers?

df[df['Total'] == df['Total'].max()]
```

```python
YearMaxCrime = df[df['Total'] == df['Total'].max()]
```

```python
#So 1991 had the highest number of crimes

YearMaxCrime.Year
```

```python
#Which year had the highest number of rapes?

YearMaxRapes = df[df['Forcible_Rape'] == df['Forcible_Rape'].max()]
```

```python
#So 1992 had the max number of rapes

YearMaxRapes.Year
```

```python
# Which year had the highest per capita crime?
```

```python
# create a new column 'PerCapitaCrime'

df['PerCapitaCrime'] = (df['Total'] / df['Population'])*100
```

```python
df
```

```python
#Which year had the highest per capita crime?

YearMaxPCC = df[df['PerCapitaCrime'] == df['PerCapitaCrime'].max()]
```

```python
#So year 1980 had the highest percapita crimes
YearMaxPCC.Year
```
